```python
def grad(X,Y,theta):
    m=X.shape[0]
    grad=np.zeros(2,)
    for i in range(m):
        yi=theta[1]*X[i]+theta[0]
        grad[0]+=2*(yi-Y[i])
        grad[1]+=2*(yi-Y[i])*X[i]
    return grad
    
```

```python
def gradient_descent(X,Y,learning_rate=0.001,max_steps=1000):
    theta=np.zeros((2,))
    error_list=[]
    for i in range(max_steps):
        #update theta
        g=grad(X,Y,theta)
        theta[0]=theta[0]-learning_rate*g[0]
        theta[1]=theta[1]-learning_rate*g[1]
        #Ya=theta[1]*X[i]+theta[0]
        #plt.scatter(X[i],Ya)
        #plt.plot
        
        error=total_error(X,Y,theta)
        error_list.append(error)
        #plt.show()
    return theta,error_list
    
```

```python
def total_error(X,Y,theta):
    m=X.shape[0]
    e=0
    for i in range(m):
        yi=theta[1]*X[i]+theta[0]
        e+=(yi-Y[i])**2
    return e
```
