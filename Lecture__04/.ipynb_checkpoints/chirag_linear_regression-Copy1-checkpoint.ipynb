```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
X=pd.read_csv('./Dataset/Training/Linear_X_Train.csv')
Y=pd.read_csv('./Dataset/Training/Linear_Y_Train.csv')
```

```python
X=np.random.randn(100)
e=np.random.randn(100)
Y=X*2-5-e
```

```python
theta=np.zeros((2,1))
#theta[0]=slope
#theta[1]=intercept
```

```python
def hypothesis(X,theta):
    return theta[0]*X+theta[1]
```

```python

def error(Y,h):
    sum=0
    for i in range(X.shape[0]):
        sum+=(Y[i]-h[i])**2
    return sum
    
```

```python
#gradient_descent

def grad(x,y,theta):
    grad1=np.zeros((2,1),dtype=float)
    for i in range(X.shape[0]):
        grad1[0]+=2*(theta[0]*x[i]+theta[1]-y[i])*x[i]
        grad1[1]+=2*(theta[0]*x[i]+theta[1]-y[i])
    return grad1
```

```python
#theta update
def theta_update(X,Y,learning_rate=0.001,steps=1000):
    e_list=[]
    for i in range(steps):
        grad1=grad(X,Y,theta)
        theta[0]=theta[0]-learning_rate*grad1[0]
        theta[1]=theta[1]-learning_rate*grad1[1]

        #print(theta)
        h=hypothesis(X,theta)
        e1=error(Y,h)
        e_list.append(e1)
    return theta,e_list
```
