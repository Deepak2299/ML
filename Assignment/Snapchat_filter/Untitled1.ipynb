Converting Bgr to Bgra

```python
Test=cv2.cvtColor(Test,cv2.COLOR_BGR2BGRA)
```

Loading the haarcascades


```python
eye_detect=cv2.CascadeClassifier('./third-party/frontalEyes35x16.xml')
nose_detect=cv2.CascadeClassifier('./third-party/Nose18x15.xml')
```

```python
eyes=eye_detect.detectMultiScale(Test,1.3,5)
nose=nose_detect.detectMultiScale(Test,1.3,5)
```

```python
x,y,w,h=eyes[0]
x1,y1,h1,w2=nose[0]
```

```python
cv2.rectangle(Test,(x,y),(x+h,y+w),(255,0,0),3)
cv2.rectangle(Test,(x1,y1),(x1+w1,y1+h1),(255,0,0),3)
plt.imshow(Test)
plt.show()
```

```python
eye_section=Test[y:y+w,x:x+h]

nose_section=Test[y1:y1+w1,x1:x1+h1]
```
