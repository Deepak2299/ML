```python
ones=np.ones((99,1))
X=np.hstack((ones,X))
```

```python
Y.shape
```

```python
def getW(X,X_Test,tau):
    m=X.shape[0]
    w=np.zeros((m,m))

    for i in range(m):
        val=np.exp(-np.dot((X[i]-X_Test).T,(X[i]-X_Test))/(2*tau*tau))
        w[i,i]=val
    return w
```

```python
Y=Y.reshape((-1,1))
```

```python
X_t=np.linspace(-5,12.5,99)
inf=float("inf")
n_inf=float("-inf")
```

```python
inf
```

```python
tauuu=[-1,1,2,3,4,inf,n_inf]
for i in tauuu:
    Pred_Y=[]
    for x in X_t:
        w=getW(X,x,i)
        theta = np.dot(np.linalg.pinv(np.dot(np.dot(X.T,w),X)),np.dot(np.dot(X.T,w),Y))
        value=theta[1,0]*x+theta[0,0]
        Pred_Y=list(Pred_Y)
        Pred_Y.append(value)
        Pred_Y=np.array(Pred_Y)
    plt.scatter(Z,Y,color="blue")
    plt.scatter(X_t,Pred_Y,label=i,color="red")
    
    plt.legend()
    plt.show()
    
```
