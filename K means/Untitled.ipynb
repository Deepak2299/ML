```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python

img=cv2.imread('/home/chirag/Desktop/messi5.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show
```

```python
img.shape
```

```python
X=img.reshape((-1,3))
```

```python
plt.imshow(img)
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(X)
```

```python
A=kmeans.cluster_centers_
j=kmeans.cluster_centers_.shape
```

```python
colors=[]
for i,color in enumerate(A):
    plt.subplot(1,100,i+1)
    colors.append(color)
    squares=np.zeros((100,100,3),dtype="uint8")
    squares[:,:,:]=color
    plt.imshow(squares)
    plt.axis("off")
```

```python
X_NEW=X
for i in range(X.shape[0]):
    idx=kmeans.labels_[i]
    B=colors[idx]
    X_NEW[i]=B
```

```python
X=X.reshape(img.shape)
X_NEW=X_NEW.reshape(img.shape)
C=np.ones((img.shape[0],10,3))
```

```python
B=np.hstack((img/255,C,X_NEW/255))
plt.figure(figsize=(20,20))
plt.title("Before and After :}")
plt.imshow(B)
plt.axis("off")
plt.show()

```
